```python
#The accompanying codebook 'MIG Codebook' is included in this github repository for reference

import pandas as pd

#Load the dataset in and filter columns relevant to the investigation (legal status, occupation, wages, wage frequency, health etc.)
df = pd.read_csv('/content/drive/MyDrive/mig174.csv')
df1 = df.filter(['sex','yrborn','usyrl','usdocl','usstatel','usoccl','usbyl','hrwage','hrweek','ushealth','mxhealth'])
df1 #Take a look at what we're working with - 9052 rows
```

```python
#Drop data entries before 1980 (arbitrary cutoff)
df2 = df1.drop(df1[(df1['usyrl'] < 1980)].index)

#Drop N/A or miscellaneous data entries
df3 = df2.drop(df2[(df2['usyrl'] == 9999)].index)
df4 = df3.drop(df3[(df3.usdocl == 9999)].index)

#Recode 'usdocl' to two groups - Undocumented and Legal statuses
def usdocl_recode(usdocl):
    if usdocl == 8:
        return 'Undocumented'
    else:
        return 'Legal'

#Apply recode to dataframe df4 under new column 'Immigrant Legal Status'
df4['Immigrant Legal Status'] = df4['usdocl'].apply(usdocl_recode)

df5 = df4.drop(df4[(df4.usbyl == 8888)].index)
df6 = df5.drop(df5[(df5.usbyl == 9999)].index)

#Similar recode method for wage payment frequency
def usbyl_recode(usbyl):
  if usbyl == 1:
    return ('Hourly')
  if usbyl == 2:
    return ('Daily')
  if usbyl == 3:
    return ('Weekly')
  if usbyl == 4:
    return ('Biweekly')
  if usbyl == 5:
    return ('Monthly')
  if usbyl == 6:
    return ('Yearly')

df6['Wage Frequency'] = df6['usbyl'].apply(usbyl_recode)

#Dropping more N/A or miscellaneous data entries
df7 = df6.drop(df6[(df6.hrwage == 9999)].index)
df8 = df7.drop(df7[(df7.hrweek == 8888)].index)
df9 = df7.drop(df8[(df8.hrweek == 9999)].index)

def sex_recode(sex):
  if sex == 1:
    return 'Male'
  if sex == 2:
    return 'Female'

df9['Sex'] = df9['sex'].apply(sex_recode)
```

```python
#Recode 'usoccl' to reflect occupation classifications according to the International Standard Classification of Occupations found here (https://ilostat.ilo.org/resources/concepts-and-definitions/classification-occupation/)
#Occupation codes are found in Apendix D in the github repository 

def usoccl_recode(usocc1):
    if usocc1 in range (10,99):
        return ('Not in Labor Force')
    if usocc1 in range(110,119):
        return ('Professionals')
    if usocc1 in range (120,129):
        return ('Technicians and Associate Professionals')
    if usocc1 in range (130,139):
        return ('Professionals')
    if usocc1 in range(140,149):
        return ('Arts, Performances, Sports')
    if usocc1 in range (210,219):
        return ('Managers')
    if usocc1 in range (410,419):
        return ('Skilled Agricultural, Forestry, Fishery')
    if usocc1 in range (510,519):
        return ('Managers')
    if usocc1 in range (520,529):
        return ('Managers')
    if usocc1 in range (530,539):
        return ('Machine Operators and Assemblers')
    if usocc1 in range (540,549):
        return ('Machine Operators and Assemblers')
    if usocc1 in range (550,559):
        return ('Machine Operators and Assemblers')
    if usocc1 in range (610,619):
        return ('Clerical Support')
    if usocc1 in range (620,629):
        return ('Administrative and Support')
    if usocc1 in range (710,719):
        return ('Service and Sales')
    if usocc1 in range (720,729):
        return ('Domestic and Personal Service')
    if usocc1 in range (810,819):
        return ('Domestic and Personal Service')
    if usocc1 == 820:
        return ('Domestic and Personal Service')
    if usocc1 in range (830,839):
        return ('Not in Labor Force')
    else:
        return ('Not in Labor Force')

df9['US Occupation'] = df9['usoccl'].apply(usoccl_recode)

df10 = df9.drop(df9[(df9['US Occupation'] == 'Not in Labor Force')].index)
```

```python
#Determine the number of data entries for each state to get an idea of the spread of immigrants
#Refer to Appendix C to get State Codes
states = df10.groupby(['usstatel']).size().sort_values(ascending = False).reset_index(name = 'count')
states
```

```python
#Based on the table above, we're going to include only entries from California as it has the highest number and is hopefully more representative and will yield more targetted interpretations
df11 = df10.drop(df10[(df10.usstatel != 105)].index)

def usstatel_recode(usstatel):
  if usstatel == 105:
    return 'California'

df11['usstatel_2'] = df11['usstatel'].apply(usstatel_recode)

#Drop all old columns from the original dataset that has since been cleaned 
Final_1 = df11.drop(['sex','usdocl', 'usstatel','usoccl','usbyl','ushealth','mxhealth'], axis=1)
Final_1

#Renaming columns
Final_1.columns = ['Year Born','Year Last Migrated','Hourly Wage','Hours Worked Weekly','Immigrant Legal Status','Wage Payment Frequency','Sex','US Occupation','US State']
Final_1 #Left with 1463 rows after data is cleaned 
```

```python
#Examine demographics of California's immmigrant population by creating a dataframe of specific columns

dem = Final_1.groupby(['Immigrant Legal Status','Sex'])['Sex'].agg(['count']).reset_index()
dem
```

```python
#Visualize this data in a sunburst chart
import plotly.express as px

dem_fig = px.sunburst(dem, path=['Immigrant Legal Status', 'Sex'], values='count', color='Immigrant Legal Status', title='Demographic Breakdown of Mexican Immigrants in California', color_discrete_map={'Legal':'#a50026', 'Undocumented':'#313695'})
dem_fig
```

```python
#Make 'Legal' and 'Undocumented' columns by pivoting the dataframe, so that we can calculate percentage of each sex for both legal statuses

dem_pivot = dem.pivot(index='Sex', columns='Immigrant Legal Status', values='count').reset_index()
dem_pivot['Percent Legal'] = (dem_pivot['Legal'] / dem_pivot['Legal'].sum()) * 100
dem_pivot['Percent Undocumented'] = (dem_pivot['Undocumented'] / dem_pivot['Undocumented'].sum()) * 100
dem_pivot
```

```python
#Compare occupations by legal status
#First, group the relevant data
ocp = Final_1.groupby(['US Occupation','Immigrant Legal Status'])['Immigrant Legal Status'].agg(['count']).reset_index()
ocp

ocp_pivot = ocp.pivot(index='US Occupation', columns='Immigrant Legal Status', values='count').reset_index()
ocp_pivot
```

```python
#Convert count to percentage
ocp_pivot['Percent Legal'] = (ocp_pivot['Legal'] / ocp_pivot['Legal'].sum()) * 100
ocp_pivot['Percent Undocumented'] = (ocp_pivot['Undocumented'] / ocp_pivot['Undocumented'].sum()) * 100
ocp_pivot
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create subplots for 'Legal' and 'Undocumented' respectively
ocp_pivot_fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])
ocp_pivot_fig.add_trace(go.Pie(labels=ocp_pivot['US Occupation'], values= ocp_pivot['Percent Legal'], sort=False, scalegroup='one', name="Legal Immigrants"),
              1, 1,)
ocp_pivot_fig.add_trace(go.Pie(labels=ocp_pivot['US Occupation'], values= ocp_pivot['Percent Undocumented'], sort=False, scalegroup='one', name="Undocumented Immigrants"),
              1, 2)

# Use `hole` to create a donut-like pie chart
colors = ['#d73027','#f46d43','#fdae61','#fee090','#ffffbf','#e0f3f8','#abd9e9','#74add1','#4575b4']
ocp_pivot_fig.update_traces(hole=.5, hoverinfo="label+percent", marker=dict(colors=colors))

ocp_pivot_fig.update_layout(
    title_text="Occupation Types by Immigrant Legal Status",
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='Legal', x=0.195, y=0.5, font_size=20, showarrow=False),
                 dict(text='Undocumented', x=0.853, y=0.5, font_size=20, showarrow=False)])

ocp_pivot_fig

#order it the same way color wise
```

```python
#Compare mean hourly wages by legal status
wages = Final_1.groupby(['US Occupation','Immigrant Legal Status'])['Hourly Wage'].agg(['mean']).reset_index()
wages.columns = ['US Occupation','Immigrant Legal Status','Mean Hourly Wage']

wages_pivot = wages.pivot(index='US Occupation', columns='Immigrant Legal Status', values='Mean Hourly Wage').reset_index()

#Find wage difference by deducting 'Legal' wages from 'Undocumented' wages
wages_pivot['Wage Difference'] = wages_pivot['Undocumented'] - wages_pivot['Legal']
wages_pivot
```

```python
#I intentionally did not drop rows for 'ushealth' and 'mxhealth' because health data was only collected in later studies. So we're going to use the first dataset for all visualizations not associated with health, and a different, smaller dataset will be used for investigations into health.
#The entries for health are not standardized so I have recoded them individually and applied them to the dataset as new columns as shown

def ushealth_recode(ushealth):
  if ushealth == 1:
    return 1
  if ushealth == ' 1':
    return 1
  if ushealth == 2:
    return 2
  if ushealth == ' 2':
    return 2
  if ushealth == 3:
    return 3
  if ushealth == ' 3':
    return 3
  if ushealth == 4:
    return 4
  if ushealth == ' 4':
    return 4
  else:
    return 0

df11['Self-rated Health Pre-migration'] = df11['ushealth'].apply(ushealth_recode)

def mxhealth_recode(mxhealth):
  if mxhealth == 1:
    return 1
  if mxhealth == ' 1':
    return 1
  if mxhealth == 2:
    return 2
  if mxhealth == ' 2':
    return 2
  if mxhealth == 3:
    return 3
  if mxhealth == ' 3':
    return 3
  if mxhealth == 4:
    return 4
  if mxhealth == ' 4':
    return 4
  else:
    return 0
 
df11['Self-rated Health Post-migration'] = df11['mxhealth'].apply(mxhealth_recode)

#Dropping irrelevant entries
df12 = df11.drop(df11[(df11['Self-rated Health Pre-migration'] == 0)].index)
df13 = df12.drop(df12[(df12['Self-rated Health Post-migration'] == 0)].index)
```

```python
Final_2 = df13.drop(['sex','usdocl', 'usstatel','usoccl','usbyl'], axis=1)
Final_2
#From Final_1 with 1463 entries, cleaning the health data has reduced the dataset to only 274 entries
```

```python
#Calculate the change in self-rated health pre- and post-migration
Final_2['Change'] = Final_2['Self-rated Health Pre-migration'] - Final_2['Self-rated Health Post-migration']

#Assign labels to calculations
def recode(Change):
  if Change == -1:
    return 'Declined'
  if Change == 0:
    return 'Maintained'
  else:
    return 'Improved'

Final_2['Label'] = Final_2['Change'].apply(recode)

health = Final_2.groupby(['Immigrant Legal Status','Label'])['Label'].agg(['count']).reset_index()
health
```

```python
health_pivot = health.pivot(index='Label', columns='Immigrant Legal Status', values='count').reset_index()
health_pivot
```

```python
#Calculate percentage
health_pivot['Percent Legal'] = (health_pivot['Legal'] / health_pivot['Legal'].sum()) * 100
health_pivot['Percent Undocumented'] = (health_pivot['Undocumented'] / health_pivot['Undocumented'].sum()) * 100
health_pivot
```

```python
#Create individual barplots for 'Legal' and 'Undocumented'
health_pivot_fig = make_subplots(rows=1, cols=2, shared_yaxes=True,subplot_titles=("Legal","Undocumented"))
colors = ['#f46d43','#fee090','#4575b4']
health_pivot_fig.add_trace(go.Bar(x=health_pivot['Label'], y=health_pivot['Percent Legal'], name="Legal", marker_color=colors),1, 1)
health_pivot_fig.add_trace(go.Bar(x=health_pivot['Label'], y=health_pivot['Percent Undocumented'], name="Undocumented", marker_color=colors),1, 2)

health_pivot_fig.update_layout(showlegend=False, title_text="Post-Migration Self-Rated Health Status Change by Immigrant Legal Status")
health_pivot_fig.update_traces(hoverinfo="x+y")
health_pivot_fig.update_yaxes(title='Percentage %')
health_pivot_fig
```
